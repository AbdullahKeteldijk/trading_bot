```python
def strip_ticker(filename):
    pattern = "_(.*?)-"
    return re.search(pattern, filename).group(1)
    
def strip_exchange(filename):
    pattern = "(.*?)_"
    return re.search(pattern, filename).group(1)
```

```python
def concat_csv(path):
    
    columns = ["Date", "Open", "High", "Low", "Close", "Volume", "Ticker"]
    df = pd.DataFrame(columns=columns)
    
    for filename in os.listdir(path):
        if ".csv" in filename:
            ticker = strip_ticker(filename)
            exchange = strip_exchange(filename)
            df_temp = pd.read_csv(f"{path}/{filename}")
            df_temp["Ticker"] = ticker
            df_temp["Exchange"] = exchange
            df = pd.concat([df, df_temp])
#             df["Date"] = pd.to_datetime(df["Date"], unit='s')

    return df
```

```python
def save_ticker(df, path):

    unique_tickers = df["Ticker"].unique()
    
    for ticker in unique_tickers:
        
        df_unique = df[df['Ticker'] == ticker]
        df_dedup = df_unique.drop_duplicates(subset = ["Date"])
        df_dedup["Date"] = pd.to_datetime(df_dedup["Date"], unit='ms')
        
        filename = f"{ticker}-USD.csv"
        df_dedup.to_csv(f"{path}/{filename}", index=False)

```

```python
path_api_data = "data/request_1h"
path_save_data = "data/one_hour"

df = concat_csv(path_api_data)
df
```
