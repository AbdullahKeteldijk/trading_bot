```python
import json
import ccxt
import pandas as pd
import numpy as np
```

```python
config = json.load(open('../logger_config.json'))
config
exchange = ccxt.binance({
    "apiKey": config["BINANCE_API_KEY"],
    "secret": config["BINANCE_SECRET_KEY"]
})
```

```python
balance = exchange.fetch_balance()
tickers = exchange.fetch_tickers()
```

```python
portfolio = pd.DataFrame(balance["free"].items(), columns=["token", "amount"]) #.sort_values("amount", ascending=False).iloc[:12]
value_list = []

for ticker, amount in portfolio.values:
    if ticker == "BUSD":
        value = amount
    else:
        try:
            value = tickers[f"{ticker}/BUSD"]["ask"] * amount
        except:
            value = 0
    value_list.append(value)

portfolio["value"] = value_list    
```

```python
import datetime


df_portfolio = portfolio.sort_values("value", ascending=False).iloc[:11]

df_portfolio["Date"] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
# df_portfolio["Date"] = pd.to_datetime(df_portfolio["Date"])

date = datetime.datetime.now().date()
df_portfolio.to_csv("../data/logger/test.csv", index=False)

```

```python
def fetch_portfolio(exchange):

    balance = exchange.fetch_balance()
    tickers = exchange.fetch_tickers()
    date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    portfolio = pd.DataFrame(balance["free"].items(), columns=["token", "amount"]) 
    value_list = []

    for ticker, amount in portfolio.values:
        if ticker == "BUSD":
            value = amount
        else:
            try:
                value = tickers[f"{ticker}/BUSD"]["ask"] * amount
            except:
                value = 0
        value_list.append(value)

    portfolio["value"] = value_list    
    portfolio["value"] = date
    portfolio = portfolio.sort_values("value", ascending=False).iloc[:11]
    portfolio.to_csv(f"../data/logger/portfolio-template-{date}.csv", index=False)
    


```

```python
total = balance["total"]
ticker = exchange.fetch_tickers()

data_dict = {}

for key, value in total.items():
    try:
        data_dict[f"{key}/BUSD"] = [value, ticker[f"{key}/BUSD"]["ask"]]
    except Exception as e:
        print(f"Error could not match: {e}")

data_dict
```

```python
df_keys = pd.DataFrame(data_dict.keys(), columns=["pair"])
df_values = pd.DataFrame(data_dict.values(), columns=["holdings", "price"])
df_portfolio = pd.concat([df_keys, df_values], axis=1)
df_portfolio["value"] = df_portfolio["holdings"] * df_portfolio["price"]
df_portfolio = df_portfolio.sort_values("value", ascending=False)
df_portfolio
```
